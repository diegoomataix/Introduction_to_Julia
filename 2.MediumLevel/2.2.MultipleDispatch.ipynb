Depending on the arguments passed to the function, the function will use a different method.

```julia
methods(+) # To see how many existing methods are available for the + operator
```

It is possible to use the @which macro to learn which method is being used

```julia
@which 3+3
```

```julia
@which 3.0 + 3.0
```

```julia
@which 3 + 3.0
```

It is possible to create new methods

```julia
import Base: +
```

```julia
"hello " + "world!"
```

```julia
+(x::String, y::String) = string(x,y)
```

```julia
"hello " + "world!"
```

```julia
@which "hello " + "world!"
```

```julia
foo(x, y) = println("duck-typed foo!")
foo(x::Int, y::Float64) = println("foo with an integer and a float!")
foo(x::Float64, y::Float64) = println("foo with two floats!")
foo(x::Int, y::Int) = println("foo with two integers")
```

```julia
foo(1,1)
```

```julia
foo(1, 1.0)
```

```julia
foo(1.,1.)
```

```julia
foo(true, false) # Since no boolean dispatch is available, the generic function dispatch is called
```
