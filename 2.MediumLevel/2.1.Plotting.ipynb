```julia
using Plots
```

```julia
x = -3:0.1:3
f(x) = x^2

y = f.(x)
```

```julia
gr() # Load the GR backend
```

```julia
plot(x, y, label="line: f(x)=x^2")
scatter!(x, y, label="points: f(x)=x^2") # The ! is needed to plot both graphs in the same figure
```

```julia
plotlyjs() # Load the PlotlyJS backend
```

```julia
plot(x, y, label="line: f(x)=x^2")
scatter!(x, y, label="points: f(x)=x^2") 
```

```julia
globaltemperatures = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
numpirates = [45000, 20000, 15000, 5000, 400, 17]
```

```julia
# Plot the data
plot(numpirates, globaltemperatures, legend=false)
scatter!(numpirates, globaltemperatures, legend=false)

# This reverses x axis
xflip!()

# Add titles and labels
xlabel!("Number of pirates")
ylabel!("Global temperature")
title!("Global temperature vs. number of pirates")
```
